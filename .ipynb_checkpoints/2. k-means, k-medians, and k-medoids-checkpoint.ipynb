### Internal Validation

```python
def pooled_within_ssd(X, y, centroids, dist):
    """Compute pooled within-cluster sum of squares around the cluster mean
    
    Parameters
    ----------
    X : array
        Design matrix with each row corresponding to a point
    y : array
        Class label of each point
    centroids : array
        Number of pairs to sample
    dist : callable
        Distance between two points. It should accept two arrays, each 
        corresponding to the coordinates of each point
        
    Returns
    -------
    float
        Pooled within-cluster sum of squares around the cluster mean
    """
    ### BEGIN SOLUTION
    ni = np.bincount(y.astype(int))
    return sum(dist(x_i, centroids[y_i])**2 / (2*ni[y_i]) 
                  for x_i, y_i in zip(X, y.astype(int)))
    ### END SOLUTION
```

```python
def gap_statistic(X, y, centroids, dist, b, clusterer, random_state=None):
    """Compute the gap statistic
    
    Parameters
    ----------
    X : array
        Design matrix with each row corresponding to a point
    y : array
        Class label of each point
    centroids : array
        Number of pairs to sample
    dist : callable
        Distance between two points. It should accept two arrays, each 
        corresponding to the coordinates of each point
    b : int
        Number of realizations for the reference distribution
    clusterer : KMeans
        Clusterer object that will be used for clustering the reference 
        realizations
    random_state : int, default=None
        Determines random number generation for realizations
        
    Returns
    -------
    gs : float
        Gap statistic
    gs_std : float
        Standard deviation of gap statistic
    """
    rng = np.random.default_rng(random_state)
    ### BEGIN SOLUTION
    from sklearn.base import clone
    mins = X.min(axis=0)
    maxs = X.max(axis=0)
    log_ws = []
    for _ in range(b):
        clusterer_i = clone(clusterer)
        clusterer_i.set_params(n_clusters=len(centroids))
        X_ref = rng.uniform(mins, maxs, size=X.shape)
        y_ref = clusterer_i.fit_predict(X_ref)
        log_ws.append(np.log(pooled_within_ssd(
            X_ref, y_ref, clusterer_i.cluster_centers_, dist)))
    return (np.mean(log_ws - 
                    np.log(pooled_within_ssd(X, y, centroids, dist))),
            np.std(log_ws - np.log(pooled_within_ssd(X, y, centroids, dist))))
    ### END SOLUTION
```

### External Validation

```python
from sklearn.metrics import confusion_matrix
def purity(y_true, y_pred):
    """Compute the class purity
    
    Parameters
    ----------
    y_true : array
        List of ground-truth labels
    y_pred : array
        Cluster labels
        
    Returns
    -------
    purity : float
        Class purity
    """
    ### BEGIN SOLUTION
    m = confusion_matrix(y_true, y_pred)
    return m.max(axis=0).sum() / np.sum(m)
    ### END SOLUTION
```

```python
def cluster_range(X, clusterer, k_start, k_stop, actual=None):
    ys = []
    inertias = []
    chs = []
    scs = []
    gss = []
    gssds = []
    ps = []
    amis = []
    ars = []
    for k in range(k_start, k_stop+1):
        clusterer_k = clone(clusterer)
        ### BEGIN SOLUTION
        clusterer_k.set_params(n_clusters=k)
        y = clusterer_k.fit_predict(X)
        ys.append(y)
        inertias.append(clusterer_k.inertia_)
        chs.append(calinski_harabasz_score(X, y))
        scs.append(silhouette_score(X, y))
        if actual is not None:
            ps.append(purity(actual, y))
            amis.append(adjusted_mutual_info_score(actual, y))
            ars.append(adjusted_rand_score(actual, y))
        ### END SOLUTION
        gs = gap_statistic(X, y, clusterer_k.cluster_centers_, 
                                 euclidean, 5, 
                                 clone(clusterer).set_params(n_clusters=k), 
                                 random_state=1337)
        gss.append(gs[0])
        gssds.append(gs[1])
    ### BEGIN SOLUTION
    res = {
        'ys': ys,
        'inertias': inertias,
        'chs': chs,
        'gss': gss,
        'gssds': gssds,
        'scs': scs
    }
    if actual is not None:
        res['ps'] = ps
        res['amis'] = amis
        res['ars'] = ars
    return res
    ### END SOLUTION
```

```python
def plot_clusters(X, ys):
    """Plot clusters given the design matrix and cluster labels"""
    k_max = len(ys) + 1
    k_mid = k_max//2 + 2
    fig, ax = plt.subplots(2, k_max//2, dpi=150, sharex=True, sharey=True, 
                           figsize=(7,4), subplot_kw=dict(aspect='equal'),
                           gridspec_kw=dict(wspace=0.01))
    for k,y in zip(range(2, k_max+1), ys):
        if k < k_mid:
            ax[0][k%k_mid-2].scatter(*zip(*X), c=y, s=1, alpha=0.8)
            ax[0][k%k_mid-2].set_title('$k=%d$'%k)
        else:
            ax[1][k%k_mid].scatter(*zip(*X), c=y, s=1, alpha=0.8)
            ax[1][k%k_mid].set_title('$k=%d$'%k)
    return ax
```

```python
def plot_internal(inertias, chs, scs, gss, gssds):
    """Plot internal validation values"""
    fig, ax = plt.subplots()
    ks = np.arange(2, len(inertias)+2)
    ax.plot(ks, inertias, '-o', label='SSE')
    ax.plot(ks, chs, '-ro', label='CH')
    ax.set_xlabel('$k$')
    ax.set_ylabel('SSE/CH')
    lines, labels = ax.get_legend_handles_labels()
    ax2 = ax.twinx()
    ax2.errorbar(ks, gss, gssds, fmt='-go', label='Gap statistic')
    ax2.plot(ks, scs, '-ko', label='Silhouette coefficient')
    ax2.set_ylabel('Gap statistic/Silhouette')
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax2.legend(lines+lines2, labels+labels2)
    return ax
```

```python
def plot_external(ps, amis, ars):
    """Plot external validation values"""
    fig, ax = plt.subplots()
    ks = np.arange(2, len(ps)+2)
    ax.plot(ks, ps, '-o', label='PS')
    ax.plot(ks, amis, '-ro', label='AMI')
    ax.plot(ks, ars, '-go', label='AR')
    ax.set_xlabel('$k$')
    ax.set_ylabel('PS/AMI/AR')
    ax.legend()
    return ax
```

### Some measures for k-medians


```python
def gap_statistic_kmedians(X, y, centroids, b):
    """Compute the gap statistic for a k-medians clusterer
    
    Parameters
    ----------
    X : array
        Design matrix with each row corresponding to a point
    y : array
        Class label of each point
    centroids : array
        Number of pairs to sample
    b : int
        Number of realizations for the reference distribution
        
    Returns
    -------
    gs : float
        Gap statistic
    gs_std : float
        Standard deviation of gap statistic
    """
    np.random.seed(1337)
    ### BEGIN SOLUTION
    mins = X.min(axis=0)
    maxs = X.max(axis=0)
    log_ws = []
    for _ in range(b):
        X_ref = np.random.uniform(mins, maxs, size=X.shape)
        kmd_i = kmedians(X_ref, X_ref[:len(centroids),:], ccore=False)
        kmd_i.process()
        y_ref = np.zeros_like(y)
        for cluster, points in enumerate(kmd_i.get_clusters()):
            y_ref[points] = cluster
        log_ws.append(np.log(pooled_within_ssd(
            X_ref, y_ref, kmd_i.get_medians(), cityblock)))
    return (np.mean(log_ws - 
                    np.log(pooled_within_ssd(X, y, centroids, cityblock))),
            np.std(log_ws - 
                   np.log(pooled_within_ssd(X, y, centroids, cityblock))))
    ### END SOLUTION
```

```python
def cluster_range_kmedians(X, k_start, k_stop, actual=None):
    ys = []
    inertias = []
    chs = []
    scs = []
    gss = []
    gssds = []
    ps = []
    amis = []
    ars = []
    for k in range(k_start, k_stop+1):
        clusterer_k = kmedians(X, X[:k,:], ccore=False)
        ### BEGIN SOLUTION
        clusterer_k.process()
        clusters = clusterer_k.get_clusters()
        y = np.zeros(len(X))
        for cluster, point in enumerate(clusters):
            y[point] = cluster
        ys.append(y)
        
        centers = clusterer_k.get_medians()
        inertias.append(sum([sum([cityblock(X[point], center)**2 
                                  for point in points]) 
                           for points, center in zip(clusters, centers)]))
        
        chs.append(calinski_harabasz_score(X, y))
        scs.append(silhouette_score(X, y))
        if actual is not None:
            ps.append(purity(actual, y))
            amis.append(adjusted_mutual_info_score(actual, y))
            ars.append(adjusted_rand_score(actual, y))
        ### END SOLUTION
        gs = gap_statistic_kmedians(X, y, clusterer_k.get_medians(), 5)
        gss.append(gs[0])
        gssds.append(gs[1])
    ### BEGIN SOLUTION
    res = {
        'ys': ys,
        'inertias': inertias,
        'chs': chs,
        'gss': gss,
        'gssds': gssds,
        'scs': scs
    }
    if actual is not None:
        res['ps'] = ps
        res['amis'] = amis
        res['ars'] = ars
    return res
    ### END SOLUTION
```

### Some measures for k-medoids

```python
def gap_statistic_kmedoids(X, y, centroids, b):
    """Compute the gap statistic for a k-medoids clusterer
    
    Parameters
    ----------
    X : array
        Design matrix with each row corresponding to a point
    y : array
        Class label of each point
    centroids : array
        Number of pairs to sample
    b : int
        Number of realizations for the reference distribution
        
    Returns
    -------
    gs : float
        Gap statistic
    gs_std : float
        Standard deviation of gap statistic
    """
    np.random.seed(1337)
    ### BEGIN SOLUTION
    mins = X.min(axis=0)
    maxs = X.max(axis=0)
    log_ws = []
    for _ in range(b):
        X_ref = np.random.uniform(mins, maxs, size=X.shape)
        kmo_i = kmedoids(X_ref, np.arange(len(centroids)), ccore=False)
        kmo_i.process()
        y_ref = np.zeros_like(y)
        for cluster, points in enumerate(kmo_i.get_clusters()):
            y_ref[points] = cluster
        log_ws.append(np.log(pooled_within_ssd(
            X_ref, y_ref, X_ref[kmo_i.get_medoids(), :], euclidean)))
    return (np.mean(log_ws - 
                    np.log(pooled_within_ssd(X, y, centroids, euclidean))),
            np.std(log_ws - 
                   np.log(pooled_within_ssd(X, y, centroids, euclidean))))
    ### END SOLUTION
```

```python
def cluster_range_kmedoids(X, k_start, k_stop, actual=None):
    ys = []
    inertias = []
    chs = []
    scs = []
    gss = []
    gssds = []
    ps = []
    amis = []
    ars = []
    for k in range(k_start, k_stop+1):
        clusterer_k = kmedoids(X, np.arange(k), ccore=False)
        ### BEGIN SOLUTION
        clusterer_k.process()
        clusters = clusterer_k.get_clusters()
        y = np.zeros(len(X))
        for cluster, point in enumerate(clusters):
            y[point] = cluster
        ys.append(y)
        
        centers = clusterer_k.get_medoids()
        inertias.append(sum([sum([euclidean(X[point], X[center])**2 
                                  for point in points]) 
                           for points, center in zip(clusters, centers)]))
        
        chs.append(calinski_harabasz_score(X, y))
        scs.append(silhouette_score(X, y))
        if actual is not None:
            ps.append(purity(actual, y))
            amis.append(adjusted_mutual_info_score(actual, y))
            ars.append(adjusted_rand_score(actual, y))
        ### END SOLUTION
        gs = gap_statistic_kmedoids(X, y, X[centers], 5)
        gss.append(gs[0])
        gssds.append(gs[1])
    ### BEGIN SOLUTION
    res = {
        'ys': ys,
        'inertias': inertias,
        'chs': chs,
        'scs': scs,
        'gss': gss,
        'gssds': gssds
    }
    if actual is not None:
        res['ps'] = ps
        res['amis'] = amis
        res['ars'] = ars
    return res
    ### END SOLUTION
```
